*Task 1*
1.	Write a program to take a string as input and print its length.
2.	Given a string, convert it to uppercase and lowercase.
3.	Write a program to count the number of vowels in a string.
4.	Check whether a given string is a palindrome (example: madam).
5.	Given a sentence, count the number of words in it.
6.	Replace all spaces in a string with underscores (_).
7.	Write a program to reverse a string without using built-in reverse functions.

```python
#1. Write a program to take a string as input and print its length.

string = "   Learning Python is simple and Python is powerful   "
print(len(string))
```

```python
#2.	Given a string, convert it to uppercase and lowercase.

string = "   Learning Python is simple and Python is powerful   "
print(string.upper()) #uppercase
print(string.lower()) #lowercase
```

```python
#3.	Write a program to count the number of vowels in a string.

string = "   Learning Python is simple and Python is powerful   "
print(string.lower().count('a')+
      string.lower().count('e')+
      string.lower().count('i')+
      string.lower().count('o')+
      string.lower().count('u'))
```

```python
#4.	Check whether a given string is a palindrome (example: madam).

string = "Madam"
new_string = string.lower()
palindrome = new_string[::-1]
print(new_string == palindrome)
```

```python
#Given a sentence, count the number of words in it.

string = "   Learning Python is simple and Python is powerful   "
print(len(string.split()))
```

```python
#6.	Replace all spaces in a string with underscores (_).

string = "   Learning Python is simple and Python is powerful   "
print(string.replace(" ", "_"))
```

```python
#7.	Write a program to reverse a string without using built-in reverse functions.

string = "   Learning Python is simple and Python is powerful   "
print(string[::-1])
```

*Task 2*
1.	Given a string, remove leading and trailing spaces.
2.	Check whether a string starts with a given character.
3.	Find the index of the first occurrence of a character in a string.
4.	Count how many times a specific character appears in a string.
5.	Split a sentence into a list of words.
6.	Join a list of words into a single string with spaces.
7.	Replace a word in a sentence with another word

```python
#1 Given a string, remove leading and trailing spaces.

string = "   Learning Python is simple and Python is powerful   "
print(string.strip())
```

```python
#2.	Check whether a string starts with a given character.

string = "   Learning Python is simple and Python is powerful   "
print(string.startswith(' '))
```

```python
#3.	Find the index of the first occurrence of a character in a string.

string = "   Learning Python is simple and Python is powerful   "
char = "L"
print(string.find(char))
```

```python
#4.	Count how many times a specific character appears in a string.

string = "   Learning Python is simple and Python is powerful   "
specific_char = "i"
print(string.find(specific_char))
```

```python
#5.	Split a sentence into a list of words.

string = "   Learning Python is simple and Python is powerful   "
print(string.split())
```

```python
#6.	Join a list of words into a single string with spaces

list = ['Learning', 'Python', 'is', 'simple', 'and', 'Python', 'is', 'powerful']

print(' '.join(list))
```

```python
#7.	Replace a word in a sentence with another word

string = "   Learning Python is simple and Python is powerful   "

print(string.replace("powerful", "very powerful"))
```
