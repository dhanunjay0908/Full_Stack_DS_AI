```python
# Question 6: Remove key 'temp' from {'a': 1, 'b': 2, 'temp': 3, 'c': 4}

dictionary = {'a': 1, 'b': 2, 'temp': 3, 'c': 4}
remove_key = dictionary.pop('temp')
print(remove_key)
print(dictionary)
```

```python
# Question 7: Find the sum of all values in {'math': 85, 'science': 92, 'english': 78}

dictionary = {'math': 85, 'science': 92, 'english': 78}
print(sum(dictionary.values()))
```

```python
# Question 8: Create a dictionary with squares of numbers 1 to 5

dictionary = {}

for i in range(1,6):
    dictionary[i] = i*i
print(dictionary)
```

```python
# Question 9: Count frequency of each character in string 'hello'

string = 'hello'
frequency = {}
for ch in string:
    if ch in frequency:
        frequency[ch] += 1
    else:
        frequency[ch] = 1
print(frequency)
```

```python
# Question 19: Create a 3-level nested dictionary: {'company': {'department': {'employee': {'name': 'John', 'id': 123}}}}

third_level_dict = {}

third_level_dict['company'] = {}
third_level_dict['company']['department'] = {}
third_level_dict['company']['department']['employee'] = {}
third_level_dict['company']['department']['employee']['name'] = 'John'
third_level_dict['company']['department']['employee']['id'] = 123

print(third_level_dict)
```

```python
# Question 20: Access employee name from 3-level nested dictionary

print(third_level_dict['company']['department']['employee']['name'])
```

```python
# Question 33: Create a dictionary with complex numbers: {'z1': 3+4j, 'z2': 1+2j}

complex_dict = {'z1': 3+4j, 'z2': 1+2j}
```

```python
# Question 34: Find magnitude of each complex number
import math
for key, value in complex_dict.items():
    
    # method 1 manual
    # magnitude1 = {key: math.sqrt(value.real**2 + value.imag**2)} 
    # print(magnitude1)
    
    # method 2 abs() built-in
    magnitude2 = {key: abs(value)}
    print(magnitude2)
```

```python
# Question 37: Create a dictionary with range values: {'r1': range(3), 'r2': range(5)}

range_dict = {'r1' : range(3), 'r2': range(2)}
print(range_dict)
```

```python
# Question 41: Create a dictionary with iterator values

iter_dict = {x: iter(range(x)) for x in range(5)}
print(iter_dict)
```
